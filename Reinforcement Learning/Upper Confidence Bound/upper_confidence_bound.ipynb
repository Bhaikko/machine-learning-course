# Upper Confidence Bound (UCB)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing UCB

```
import math 
```

```
numOfUsers = 10000                                  # N
numOfAds = 10                                       # d

adsSelected = []
numTimesAddSelected = [0] * numOfAds                # Ni(n)
sumOfRewards = [0] * numOfAds                       # Ri(n)
totalReward = 0

for n in range(0, numOfUsers):
    ad = 0
    maxUpperBound = 0                               # Maximum of Upper Confidence bound

    for i in range(0, numOfAds):
        if (numTimesAddSelected[i] > 0):
            averageReward = sumOfRewards[i] / numTimesAddSelected[i]
            deltaI = math.sqrt(1.5 * (
                math.log(n + 1) / numTimesAddSelected[i]
            ))

            upperBound = averageReward + deltaI
            
            # confience Interval will be
            # [averageRange - deltaI, averageRange + deltaI]
        else:
            upperBound = 1e400

        if (upperBound > maxUpperBound):
            maxUpperBound = upperBound
            ad = i 
    
    adsSelected.append(ad)
    numTimesAddSelected[ad] += 1

    reward = dataset.values[n, ad]
    sumOfRewards[ad] += reward
    totalReward += reward

# print(adsSelected)
```

## Visualising the results

```
# plotting historgram of rewards for each advertisements

plt.hist(adsSelected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
