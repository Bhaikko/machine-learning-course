# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('/home/bhaikko/Machine Learning/Regression/Simple Linear Regression/Salary_Data.csv')
X = dataset.iloc[:, : -1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()                      # This builds a Linear Regression Model which is independent of data

regressor.fit(X_train, y_train)                     # This method trains the Above Linear Regression Model on the training data

                                                    # LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False) shows the training is done.
```

## Predicting the Test set results

```python
y_pred = regressor.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')                        # This plots the training data as discrete values on graph
plt.plot(X_train, regressor.predict(X_train), color = 'blue')       # This plots the trained model prediction as continous values on graph
plt.title("Salary vs Experience (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set Results

```python
plt.scatter(X_test, y_test, color = 'red')                          # This plots the test data as discrete values on graph
plt.plot(X_train, regressor.predict(X_train), color = 'blue')       # Since predicted salaries of test and training set will be on same line, 
                                                                    # we dont need to replace train and test data

plt.title("Salary vs Experience (Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
